```python
from newspaper import Article
from urllib.parse import urlparse
import nltk
#nltk.download('punkt')

import pandas as pd
import numpy as np
import sklearn as skl
import nltk as nltk
#import nltk.corenlp as nlp
import seaborn as sbn
import nltk.grammar as grm
import nltk.corpus as corpus

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.linear_model import PassiveAggressiveClassifier, LinearRegression
from sklearn.feature_extraction import stop_words
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import svm
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVR,SVC


import xgboost as xgb

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neural_network import MLPClassifier
```

```python
df=pd.read_pickle("/home/rd/Desktop/prj/cleaned_df.pkl")
```

```python
df
```

```python
x=df['clean_text']
y=df['label']
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.70)
tfidf_vector=TfidfVectorizer(stop_words='english',max_df=0.70)
tfidf_train=tfidf_vector.fit_transform(train_x)
tfidf_test=tfidf_vector.transform(test_x)
```

```python
tfidf_df = pd.DataFrame(tfidf_train)
#, columns=tfidf_vector.get_feature_names()
tfidf_df
```

```python
from sklearn.svm import SVC
clf=SVC(kernel='linear')
clf.fit(tfidf_train,train_y)
```

```python
pred_svm_y_tfidf=clf.predict(tfidf_test)
```

```python
acc_scr_tfidf=accuracy_score(test_y,pred_svm_y_tfidf)
```

```python
print(acc_scr_tfidf)
```

```python
from sklearn.pipeline import Pipeline
pipeline=Pipeline([('tfidf',TfidfVectorizer(stop_words='english',max_df=0.7)),
                  ('svm',LinearSVR())])
pipeline.fit(train_x,train_y)
```

```python
score=pipeline.score(train_x,train_y)
print('accuracy',score)
```

```python
pred=pipeline.predict(test_x)
print(classification_report(test_y,pred.round()))
```
